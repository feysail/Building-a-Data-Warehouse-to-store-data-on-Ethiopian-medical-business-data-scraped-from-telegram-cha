## Task 1 - Data scraping and collection pipeline


```python
from telethon import TelegramClient
import csv
import os
import threading
import pandas as pd
from dotenv import load_dotenv
```

```python



load_dotenv('.env')
api_id = os.getenv('TG_API_ID')
api_hash = os.getenv('TG_API_HASH')
phone = os.getenv('phone')


csv_lock = threading.Lock()

async def scrape_channel(client, channel_username, media_dir):
    entity = await client.get_entity(channel_username)
    channel_title = entity.title  

    csv_file_path = r'C:\Users\ASUS VIVO\Downloads\telegram_data_scrapped.csv'
    
    async for message in client.iter_messages(entity, limit=100):
        media_path = None
        if message.media and hasattr(message.media, 'photo'):
 
            filename = f"{channel_username}_{message.id}.jpg"
            media_path = os.path.join(media_dir, filename)
            await client.download_media(message.media, media_path)

        with csv_lock:  
            with open(csv_file_path, 'a', newline='', encoding='utf-8') as file:
                writer = csv.writer(file)
                writer.writerow([channel_title, channel_username, message.id, message.message, message.date, media_path])

async def main():

    async with TelegramClient('scraping_session', api_id, api_hash) as client:
        await client.start()
        
   
        media_dir = 'photos'
        os.makedirs(media_dir, exist_ok=True)

        with open('telegram_data.csv', 'w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow(['Channel Title', 'Channel Username', 'ID', 'Message', 'Date', 'Media Path']) 
      
            channels = [
                '@DoctorsET',
                '@CheMed123',
                '@lobelia4cosmetics',
                '@yetenaweg',
                '@EAHCI'
            ]
            
     
            for channel in channels:
                await scrape_channel(client, channel, media_dir)
                print(f"Scraped data from {channel}")


await main()
```
